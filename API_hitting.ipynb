```python
import requests
import pandas as pd
```

```python
# read the csv file
df = pd.read_csv('null_stations_after_purify.csv')
```

```python
# input parameter

address = []
display_name = []

for index, row in df.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    
    url = f'https://nominatim.openstreetmap.org/reverse?format=json&lat={latitude}&lon={longitude}'

    response = requests.get(url, verify=False)

    
    if response.status_code == 200:
        data = response.json()

        address.append(data.get('address'))
        display_name.append(data.get('display_name'))
        #station_name.append(location_name.get('amenity'))
        #print(location_name.get('amenity'))
    else:
        print('Request failed with status code:', response.status_code)
```

```python
df['address'] = address
df['display_name'] = display_name
```

```python
df.head()
df.to_csv('new_stations.csv')
```

```python
new_address = []
for index, row in df.iterrows():
    address = row['address']
    
    hn = ''
    bn = ''
    hs = ''
    rn = ''
    am = ''
    
    new_add = ''
    
    if 'historic' in address:
        hs = address.get('historic')
        new_add = new_add + ' ' + hs
    
    if 'amenity' in address: 
        am = address.get('amenity')
        new_add = new_add + ' ' + am
        
    if 'building' in address:
        bn = address.get('building')
        new_add = new_add + ' ' + bn
    
    if 'house_number' in address:
        hn = address.get('house_number')
        new_add = new_add + ' ' + hn
    
    if 'road' in address: 
        rn =  address.get('road')
        new_add = new_add + ' ' + rn
        
    new_add = new_add.strip()
    
    new_address.append(new_add)
```

```python
df['station_name'] = new_address
```

```python
new_df = df[['latitude', 'longitude', 'station_name']]
new_df.head()
```

```python
new_df.to_csv('purified_stations.csv', index=False)
```
